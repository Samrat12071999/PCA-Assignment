```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
# Import Dataset
```

```python
df=pd.read_csv("C:\\Users\\pradyumn\\Documents\\Assignment\\wine.csv")
df.head()
```

```python
df.shape
```

```python
df['Type'].value_counts()
```

```python
df1=df.iloc[:,1:]
df1
```

```python
df1.info()
```

```python
# Converting data in numpy array
df_ar=df1.values
df_ar
```

```python
# Normalizing the numerical data 
df_norm=scale(df_ar)
df_norm
```

# PCA Implementation

```python
# Applying PCA Fit Transform to dataset
pca=PCA()
wine_pca=pca.fit_transform(df_norm)
wine_pca
```

```python
# PCA Components matrix or covariance Matrix
pca.components_
```

```python
# The amount of variance that each PCA has
var=pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color='red')
```

```python
# Final Dataframe
final_df=pd.concat([df['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

```python
sns.scatterplot(data=final_df,x='PC1',y='PC2',hue='Type',s = 100)  
```

# Clustering Algorithms

# 1) Agglomerative Clustering

```python
# Import Libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(df_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(df_norm),columns=['clustaglo'])
y['clustaglo'].value_counts()
```

```python
# Adding clusters to dataset
df2=df.copy()
df2['clustersid']=hclusters.labels_
df2
```

# 2) K-Means Clustering

```python
from sklearn.cluster import KMeans
import warnings
```

```python
warnings.simplefilter(action='ignore', category=FutureWarning)
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) 
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
WCSS
```

```python
# Cluster algorithm using K=3
clusters3=KMeans(3,random_state=30).fit(df_norm)
clusters3
```
